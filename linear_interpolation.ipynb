```python
import numpy as np
import pandas as pd

def linear_interpolation(x0, x1, y0, y1, x):
    return y0 + (y1 - y0) * (x - x0) / (x1 - x0)

data = pd.read_csv("SiData.csv")
for col in data.columns:
    if not "Distance" in col:
        continue
    zero_indices = np.where(data[col] == 0.0)[0]


    for i in zero_indices:
        left_index = max([j for j in range(i) if data[col][j] != 0.0], default=0)
        right_index = min([j for j in range(i + 1, len(data[col])) if data[col][j] != 0.0], default=len(data[col]) - 1)
        interpolated_value = linear_interpolation(left_index, right_index, data[col][left_index], data[col][right_index], i)
        data[col][i] = interpolated_value
```
