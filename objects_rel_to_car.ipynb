```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("./OriginalData.csv")
df.head()
```

```python
color_gradient = np.linspace(0, 1, len(df['FirstObjectDistance_X']))
plt.scatter(df['FirstObjectDistance_X'], df['FirstObjectDistance_Y'], c=color_gradient, cmap='viridis')
colorbar = plt.colorbar()
colorbar.set_label('Data Index')
```
