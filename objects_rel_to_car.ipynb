```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("./OriginalData.csv")
for col in df.columns:
    if "Distance" in col:
        df[col]=df[col]/128
df.head()
```

```python
df[df['FirstObjectDistance_X']==0].head()
```

```python
color_gradient = np.linspace(0, 1, len(df['FirstObjectDistance_X']))
plt.scatter(df['FirstObjectDistance_Y'], df['FirstObjectDistance_X'], c=color_gradient, cmap='viridis')
colorbar = plt.colorbar()
colorbar.set_label('Data Index')
plt.savefig("FirstObject.png")
```
